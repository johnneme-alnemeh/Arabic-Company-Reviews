```python
##without stemmer
def count_neg(text):
  neg = 0
  for w in word_tokenize(text):
    if w in n:
      neg +=1
  return neg

def count_pos(text):
  pos = 0
  for w in word_tokenize(text):
    if w in p:
      pos +=1
  return pos
```

```python
def pre_process(df):
  df['pos'] = df['review_description'].apply(count_pos)
  df['neg'] = df['review_description'].apply(count_neg)
  return df
pre_process(df)
```
